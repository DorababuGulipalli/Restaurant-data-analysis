Data cleaning

```python
df.isnull().sum()
```

```python
df.columns
```

```python
#Dropping unwnated Columns

df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
#Droppping duplicate rows

df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
#Cleaning Rate column

df['rate'].unique()
```

```python
#Removing string value "NEW" and '/5' from rate

def rate_clean(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(rate_clean)
df['rate'].head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Filling the null values in the rate column with mean value

df['rate'].fillna(df['rate'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#Dropping Null values from dataframe

df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.head()
```

Now the name,online_order,book_table,rate colums are clean, moving on to other columns

```python
df.rename(columns={'listed_in(type)':'type'},inplace=True)
df.head()
```

```python
#Finding the unique values in location column

df['location'].nunique()
```

```python
df['location'].unique()
```

```python
#Finding the unique values in listed_in(city) column

df['listed_in(city)'].nunique()
```

```python
df['listed_in(city)'].unique()
```

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

```python
#Cleaning the approx_cost(for two people) column

df['approx_cost(for two people)'].unique()
```

```python
#Converting all the values in 'approx_cost(for two people)' column into float values.

def removecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(removecomma)
df['approx_cost(for two people)'].unique()
```

```python
df.head()
```

Now we also cleaned approx_cost(for two people) column moving on to rest_type column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'Others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head(10)
```

```python
#Cleaning the location column

location = df['location'].value_counts(ascending  = False)
location
```

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'Others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
#Cleaning cuisine column

cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]



def group_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'Others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(group_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['type'].value_counts()
```

The type column is clean. So with this the data cleaning is done, now moving on to data visualization

Data Visualization

```python
#Count plot for various locations.

plt.figure(figsize = (16,10))
ax = sns.countplot(df['location'])
plt.xticks(rotation=90)
```
